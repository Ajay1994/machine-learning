```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("/data/insurance.csv")
```

```python
df.head()
```

```python
plt.hist(df.charges, bins = 50, normed=True);
plt.xlabel("Charges")
plt.ylabel("Probability")
plt.title("Histogram of insurance charges")
```

```python
df.charges.plot.hist(bins = 50, normed = True)
plt.xlabel("Charges")
plt.ylabel("Probability")
plt.title("Histogram of insurance charges")
```

```python
df.charges.plot.box()
```

```python
df.describe()
```

```python
np.log(df.charges).plot.box()
```

```python
df.sex.value_counts().plot.bar()
```

```python
df.sex.value_counts().plot.pie()
```

```python
df.groupby("sex").charges.mean().sort_values(ascending = False).plot.bar()
plt.title("Average of charges by gender")
plt.ylabel("Average of charges")
plt.xlabel("Gender")
```

```python
df.plot.scatter("bmi", "charges")
```

```python
colors = np.where(df.smoker == "yes", "blue", "red")
df.plot.scatter("bmi", "charges", color = colors)
```

```python
agg = df.groupby(["region", "sex"]).charges.median().unstack()
sns.heatmap(agg)
```

```python
sns.heatmap(agg, cmap=sns.color_palette("Blues"))
```
