```python
tf.reset_default_graph()
tf.set_random_seed(1)

n_x = X_train.shape[1]
n_y = 1
n_h = 3
learning_rate = 1e-1
max_iter = 5000

X = tf.placeholder(dtype=tf.float32, shape=[None, n_x], name = "X")
y = tf.placeholder(dtype=tf.float32, shape=[None, 1], name = "y")

print("X", X)
print("y", y)

W1 = tf.Variable(initial_value= tf.truncated_normal(shape=[n_x, n_h]))
b1 = tf.Variable(tf.zeros(shape=[n_h]))
print("W1", W1)
print("b1", b1)

Z1 = tf.matmul(X, W1) + b1
print("Z1", Z1)

A1 = tf.nn.relu(Z1)

W2 = tf.Variable(initial_value= tf.truncated_normal(shape=[n_h, n_y]))
b2 = tf.Variable(tf.zeros(shape=[n_y]))
print("W2", W2)
print("b2", b2)

Z2 = tf.matmul(A1, W2) + b2
print("Z2", Z2)

y_pred = Z2

mse = tf.reduce_mean((y - y_pred) ** 2)
print("mse", mse)

optimizer = tf.train.AdamOptimizer(learning_rate).minimize(mse)

scores = []
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for i in range(max_iter):
        mse_, _ = sess.run([mse, optimizer], feed_dict={X: X_train,
                                            y: y_train.reshape(-1, 1)})
        scores.append(mse_)
        if i % (max_iter//25) == 0 or i == max_iter-1:
            print(mse_)
    test_mse_, y_pred_ = sess.run([mse, y_pred], feed_dict={X: X_test,
                                            y: y_test.reshape(-1, 1)})
    print("Test mse", test_mse_)
```

```python
tf.reset_default_graph()
tf.set_random_seed(1)

n_x = X_train.shape[1]
n_y = 10
n_h = 400
learning_rate = 1e-1
max_iter = 200

X = tf.placeholder(dtype=tf.float32, shape=[None, n_x], name = "X")
y = tf.placeholder(dtype=tf.int32, shape=[None], name = "y")

print("X", X)
print("y", y)

y_oh = tf.one_hot(y, depth=n_y)
print("y_oh", y_oh)

W1 = tf.Variable(initial_value= tf.truncated_normal(shape=[n_x, n_h]))
b1 = tf.Variable(tf.zeros(shape=[n_h]))
print("W1", W1)
print("b1", b1)

Z1 = tf.matmul(X, W1) + b1
print("Z1", Z1)

A1 = tf.nn.relu(Z1)

W2 = tf.Variable(initial_value= tf.truncated_normal(shape=[n_h, n_y]))
b2 = tf.Variable(tf.zeros(shape=[n_y]))
print("W2", W2)
print("b2", b2)

Z2 = tf.matmul(A1, W2) + b2
print("Z2", Z2)

y_pred = tf.argmax(Z2, axis=1, output_type=tf.int32)

cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=Z2, labels=y_oh))

accuracy = tf.reduce_mean(tf.cast(tf.equal(y, y_pred), tf.float32))

optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)

scores = []
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    batchable = Batchable(X_train, y_train, batch_size=128, seed=1)
    
    for i in range(max_iter):
        X_batch, y_batch = batchable.next()
        cost_, _ = sess.run([cost, optimizer], feed_dict={X: X_train,y: y_train})
        scores.append(cost_)
        if i % (max_iter//25) == 0 or i == max_iter-1:
            train_accuracy_ = sess.run(accuracy, feed_dict={X: X_train,y: y_train})
            test_accuracy_ = sess.run(accuracy, feed_dict={X: X_test,y: y_test})
            print("cost", cost_, " train", train_accuracy_, "test:", test_accuracy_)
    y_pred_, A1_, W1_, W2_ = sess.run([y_pred, A1, W1, W2], feed_dict={X: X_test,y: y_test})
pd.Series(scores).plot()
```
