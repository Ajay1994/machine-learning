```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

import keras
from keras.layers import Dense, InputLayer, Dropout

import tensorflow as tf

from time import time

pd.options.display.max_columns = 1000
pd.options.display.max_rows = 10

%matplotlib inline
```

```python
X_train = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y_train = np.array([0, 1, 1, 0])
```

```python
model = keras.Sequential()
model.add(InputLayer((2, )))
model.add(Dense(1, activation="sigmoid"))
model.compile("adam", keras.losses.binary_crossentropy, ["accuracy"])

model.fit(X_train, y_train, batch_size=1, validation_data=(X_train, y_train), epochs=100)
```

```python
tboard = keras.callbacks.TensorBoard("/tmp/tf/logs/%d" % time())

tf.set_random_seed(1)
np.random.seed(1)

adam = keras.optimizers.Adam(lr = 0.01)

model = keras.Sequential()
model.add(InputLayer((2, )))
model.add(Dense(10, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(10, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(10, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.compile(adam, keras.losses.binary_crossentropy, ["accuracy"])

model.summary()

model.fit(X_train, y_train, validation_data=(X_train, y_train), epochs=100, callbacks=[tboard])
```

```python
df_train = pd.read_csv("/data/MNIST/mnist_train.csv", header=None)
df_test= pd.read_csv("/data/MNIST/mnist_test.csv", header=None)
```

```python
df_train.shape, df_test.shape
```

```python
28 * 28
```

```python
X_train = df_train.iloc[:, 1:].values / 255
y_train = df_train.iloc[:, 0].values

X_test = df_test.iloc[:, 1:].values /255
y_test = df_test.iloc[:, 0].values
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
X_train[0]
```

```python
fig, _ = plt.subplots(5, 5, figsize = (10, 10))

for i, ax in enumerate(fig.axes):
    ax.imshow(X_train[i].reshape((28, 28)), cmap="gray")
    ax.set_title(y_train[i])
    
    
plt.tight_layout()
```

```python
Y_train = keras.utils.to_categorical(y_train)
Y_test = keras.utils.to_categorical(y_test)
```

```python
tf.set_random_seed(1)
np.random.seed(1)


model = keras.Sequential()
model.add(InputLayer((784,)))
model.add(Dense(10, activation="softmax"))

model.compile("adam", keras.losses.categorical_crossentropy
              , metrics=["accuracy"])

model.summary()
```

```python
tboard = keras.callbacks.TensorBoard("/tmp/tf/logs/%d" % time())
model.fit(X_train, Y_train, validation_data=(X_test, Y_test)
          , batch_size=128
          , epochs=10
          , callbacks=[tboard])
```

```python
Y_test_pred = model.predict_proba(X_test)
```

```python
Y_test_pred.shape
```

```python
pd.DataFrame(Y_test_pred)
```

```python
y_test_pred = np.argmax(Y_test_pred, axis=1)
```

```python
y_test_pred.shape
```

```python
matches = y_test == y_test_pred
```

```python
y_test[~matches][0]
```

```python
fig, _ = plt.subplots(5, 5, figsize = (10, 10))

for i, ax in enumerate(fig.axes):
    ax.imshow(X_test[~matches][i].reshape((28, 28)), cmap="gray")
    title = "t: %d, p: %d" % (y_test[~matches][i], y_test_pred[~matches][i])
    ax.set_title(title)
    
    
plt.tight_layout()
```

```python
tf.set_random_seed(1)
np.random.seed(1)


model = keras.Sequential()
model.add(InputLayer((784,)))
model.add(Dense(100, activation="relu"))
model.add(Dense(10, activation="softmax"))

model.compile("adam", keras.losses.categorical_crossentropy
              , metrics=["accuracy"])

model.summary()

tboard = keras.callbacks.TensorBoard("/tmp/tf/logs/%d" % time())
model.fit(X_train, Y_train, validation_data=(X_test, Y_test)
          , batch_size=128
          , epochs=10
          , callbacks=[tboard])
```

```python
Y_test_pred = model.predict_proba(X_test)
y_test_pred = np.argmax(Y_test_pred, axis=1)
matches = y_test == y_test_pred
fig, _ = plt.subplots(5, 5, figsize = (10, 10))

for i, ax in enumerate(fig.axes):
    ax.imshow(X_test[~matches][i].reshape((28, 28)), cmap="gray")
    title = "t: %d, p: %d" % (y_test[~matches][i], y_test_pred[~matches][i])
    ax.set_title(title)
    
    
plt.tight_layout()
```

```python
from keras.layers import Conv2D, MaxPooling2D, Flatten
```

```python
tf.set_random_seed(1)
np.random.seed(1)


model = keras.Sequential()
model.add(InputLayer((28,28, 1)))
model.add(Conv2D(32, (5, 5), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.5))
model.add(Flatten())
model.add(Dense(400, activation="relu"))
model.add(Dense(100, activation="relu"))
model.add(Dense(10, activation="softmax"))

model.compile("adam", keras.losses.categorical_crossentropy
              , metrics=["accuracy"])

model.summary()

tboard = keras.callbacks.TensorBoard("/tmp/tf/logs/%d" % time())
model.fit(X_train.reshape(-1, 28, 28, 1), Y_train
          , validation_data=(X_test.reshape(-1, 28, 28, 1), Y_test)
          , batch_size=128
          , epochs=10
          , callbacks=[tboard])

```

```python
Y_test_pred = model.predict_proba(X_test.reshape(-1, 28, 28, 1))
y_test_pred = np.argmax(Y_test_pred, axis=1)
matches = y_test == y_test_pred
fig, _ = plt.subplots(5, 5, figsize = (10, 10))

for i, ax in enumerate(fig.axes):
    ax.imshow(X_test[~matches][i].reshape((28, 28)), cmap="gray")
    title = "t: %d, p: %d" % (y_test[~matches][i], y_test_pred[~matches][i])
    ax.set_title(title)
    
    
plt.tight_layout()
```
