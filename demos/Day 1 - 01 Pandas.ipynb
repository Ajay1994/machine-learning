```python
import numpy as np
```

```python
df.groupby("gender")["charges"].agg([np.mean, len])
```

```python
df.groupby("gender").agg({"charges": np.mean, "age": np.mean})
```

```python
df.groupby("gender").agg({"charges": [np.mean, np.median], "age":  [np.mean, np.median]})
```

```python
(df
 .groupby(["gender", "region"])
 .agg({"charges": [np.mean, np.median], "age":  [np.mean, np.median]}))
```

```python
df.pivot_table(values="charges", columns="gender", index = "region", aggfunc=np.median)
```
