```python
import pandas as pd
import numpy as np
from sklearn import *
import matplotlib.pyplot as plt

from mlxtend.plotting import plot_decision_regions

%matplotlib inline
```

```python
df = pd.read_csv("/data/iris.csv")
```

```python
df.head()
```

```python
df.Species.value_counts()
```

```python
features = ["SepalLengthCm", "PetalLengthCm"]
```

```python
fig, ax = plt.subplots()
df[df.Species == "Iris-virginica"].plot.scatter(features[0], 
                                features[1], color = "red", ax = ax, label = "Iris-virginica")
df[df.Species == "Iris-setosa"].plot.scatter(features[0], 
                                features[1], color = "blue", ax = ax, label = "Iris-setosa")
df[df.Species == "Iris-versicolor"].plot.scatter(features[0], 
                                features[1], color = "green", ax = ax, label = "Iris-versicolor")
```

```python
target = "Species"
y = np.where(df[target] == "Iris-setosa", 1, 0)
X = df[features]

X = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, 
                                        test_size = 0.3, random_state = 12345)

pipe = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree=1, include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.LogisticRegression())
])
pipe.fit(X_train, y_train)
print("Accuracy on test", pipe.score(X_test, y_test))
plot_decision_regions(X.values, y, pipe)
plt.xlabel(features[0])
plt.ylabel(features[1])
plt.title("Decision region")
```
