```python
ratings = pd.read_csv("/data/movielens/ratings.csv")
ratings
```

Find top 10 movies based on highest avg rating. 

```python
ratings[ratings.movieId == 31].shape
```

```python
ratings_agg = ratings.groupby("movieId").rating.agg([np.mean, len]).query("len > 100")
ratings_agg = ratings_agg.sort_values("mean", ascending=False)
ratings_agg = ratings_agg.iloc[:10, :]
ratings_agg
```

```python
movies.merge(ratings_agg, left_on="movieId"
        , right_index=True).loc[:, ["mean", "title"]].sort_values("mean", ascending = False)
```
