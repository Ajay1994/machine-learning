```python
import pandas as pd

from sklearn import *
import numpy as np

pd.options.display.max_rows = 10
pd.options.display.max_columns = 1000
```

```python
df = pd.read_csv("/data/kaggle/house-prices/data_combined_cleaned.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df = df[~df.SalesPrice.isnull()]
del df["Id"]
```

```python
df.shape
```

```python
df.info()
```

```python
target = "SalesPrice" # name of the target variable
y = df[target] # label .. or the target 
X = df.copy() # features ... label is not included
del X[target]
X = pd.get_dummies(X, drop_first=True) # handle catergorical variables ... one hot encoding

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y
                            , test_size = 0.3, random_state = 1)

pipe = pipeline.Pipeline([
    #("poly", preprocessing.PolynomialFeatures(degree=5, include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.LinearRegression())
])
                                      
pipe.fit(X_train, y_train)

y_train_pred = pipe.predict(X_train)
y_test_pred = pipe.predict(X_test)

print("Training mse:", metrics.mean_squared_error(y_train, y_train_pred),
     "\nTest mse:", metrics.mean_squared_error(y_test, y_test_pred), 
      "\nTraining r2:", metrics.r2_score(y_train, y_train_pred),
      "\nTest r2:", metrics.r2_score(y_test, y_test_pred))
```

```python
pipe.steps[-1][-1].coef_
```

```python
target = "SalesPrice" # name of the target variable
y = df[target] # label .. or the target 
X = df.copy() # features ... label is not included
del X[target]
X = pd.get_dummies(X, drop_first=True) # handle catergorical variables ... one hot encoding

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y
                            , test_size = 0.3, random_state = 174298)

pipe = pipeline.Pipeline([
    #("poly", preprocessing.PolynomialFeatures(degree=5, include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.Lasso(alpha=350.0, max_iter= 10000
                                      , random_state=1))
])
                                      
pipe.fit(X_train, y_train)

y_train_pred = pipe.predict(X_train)
y_test_pred = pipe.predict(X_test)

print("Training mse:", metrics.mean_squared_error(y_train, y_train_pred),
     "\nTest mse:", metrics.mean_squared_error(y_test, y_test_pred), 
      "\nTraining r2:", metrics.r2_score(y_train, y_train_pred),
      "\nTest r2:", metrics.r2_score(y_test, y_test_pred))

pipe.steps[-1][-1].coef_
```

```python
result = pd.DataFrame({"feature": X.columns, "coef": pipe.steps[-1][-1].coef_})
```
