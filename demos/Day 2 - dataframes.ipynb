# Exercise: find top 10 movies
    
1. load movies.csv as movies dataframe and ratings.csv as ratings dataframe
2. from ratings dataframe find the avg and count of rating for each movieId
3. From #2, Filter out all movieId that have fewer than 100 ratings
4. Join #3 with movies dataset to put title [Hint: use pd.merge function]
5. Sort the data based on the avg rating
6. return 10 movies with higest avg rating
7. Display only movieId, title, and avg rating

```python
!ls -1 /data/movielens/
```

```python
!head /data/movielens/movies.csv
```

```python
!head /data/movielens/ratings.csv
```

```python
movies = pd.read_csv("/data/movielens/movies.csv")
movies.sample(10)
```

```python
ratings = pd.read_csv("/data/movielens/ratings.csv")
ratings.sample(10)
```

```python
agg = (ratings
       .groupby("movieId")
       .rating.agg([np.mean, len])
       .query("len>=100"))
agg
```

```python
(pd.merge(agg, 
         movies, 
         left_index=True, 
         right_on="movieId")[["mean", "movieId", "title"]]
    .sort_values("mean", ascending = False)
    .iloc[:10, :]
)
```
