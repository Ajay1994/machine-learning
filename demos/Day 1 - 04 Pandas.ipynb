```python
df["profit_rd_ratio"] = df["Profit"] / df["R&D Spend"]
df
```

```python
def find_profit_rd_ratio(row):
    return row["Profit"] / row["R&D Spend"]
    
df["profit_rd_ratio"] = df.apply(find_profit_rd_ratio, axis = 1)
df
```

```python
df["State"] = df.State.apply(lambda s: s.upper())
df
```
