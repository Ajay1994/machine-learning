```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import *
import numpy as np
from mlxtend.plotting import plot_confusion_matrix, plot_decision_regions
```

```python
df = pd.read_csv("/data/iris.csv")
df.head()
```

```python
df.Species.value_counts()
```

```python
features = ["SepalLengthCm", "PetalLengthCm"]
```

```python

fig, ax = plt.subplots(figsize = (10, 8))
df[df.Species == "Iris-versicolor"].plot.scatter(features[0], features[1], ax = ax, color = "red")
df[df.Species == "Iris-virginica"].plot.scatter(features[0], features[1], ax = ax, color = "green")
df[df.Species == "Iris-setosa"].plot.scatter(features[0], features[1], ax = ax, color = "blue")
```

```python
target = "Species"
y = np.where(df[target] == "Iris-versicolor", 1, 0)
X = df[features]

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y
                            , test_size = 0.3, random_state = 1)

pipe = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree=5, include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.LogisticRegression())
])
                                      
pipe.fit(X_train, y_train)

y_train_pred = pipe.predict(X_train)
y_test_pred = pipe.predict(X_test)

print("accuracy", metrics.accuracy_score(y_test, y_test_pred),
"\nprecision", metrics.precision_score(y_test, y_test_pred),
"\nrecall", metrics.recall_score(y_test, y_test_pred))
plot_confusion_matrix(metrics.confusion_matrix(y_test, y_test_pred))
```

```python
plt.figure(figsize= (10, 8))
plot_decision_regions(X.values, y, pipe)
plt.xlabel(features[0])
plt.ylabel(features[1])
```
